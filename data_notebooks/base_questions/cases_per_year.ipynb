# What is the total volume of requests per year or how many 311 requests is the city receiving per year?

```python
# Import libraries we need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Some global variables
years = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018','2019','2020','2021','2022','2023']
int_years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018,2019,2020,2021,2022,2023]
```

```python
# Cleaning the data a bit:

# Read in the data
df_2011 = pd.read_csv('../../data/2011.csv', low_memory=False)
df_2012 = pd.read_csv('../../data/2012.csv', low_memory=False)
df_2013 = pd.read_csv('../../data/2013.csv', low_memory=False)
df_2014 = pd.read_csv('../../data/2014.csv', low_memory=False)
df_2015 = pd.read_csv('../../data/2015.csv', low_memory=False)
df_2016 = pd.read_csv('../../data/2016.csv', low_memory=False)
df_2017 = pd.read_csv('../../data/2017.csv', low_memory=False)
df_2018 = pd.read_csv('../../data/2018.csv', low_memory=False)
df_2019 = pd.read_csv('../../data/2019.csv', low_memory=False)
df_2020 = pd.read_csv('../../data/2020.csv', low_memory=False)
df_2021 = pd.read_csv('../../data/2021.csv', low_memory=False)
df_2022 = pd.read_csv('../../data/2022.csv', low_memory=False)
df_2023 = pd.read_csv('../../data/2023.csv', low_memory=False)
```

## The amount of cases per year

```python
# Create a list of the dataframes
dfs = [df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018,df_2019,df_2020,df_2021,df_2022,df_2023]

# Create a list of the amount of cases per year
cases_per_year = []
for i, df in enumerate(dfs):
    cases_per_year.append(len(df))
    print(f"The amount of cases for the year {int_years[i]} is " + str(len(df)) + " cases")

# Plot the amount of cases per year
plt.figure(figsize=(20,10))
plt.bar(years, cases_per_year)
plt.xlabel('Year')
plt.ylabel('Amount of cases')
plt.title('Amount of cases per year')
plt.show()
```

The data shows the cases per year from 2011-2023. There is an increase of cases from 2011 to 2018 and after a slight decrease from 2018 to 2020 to then an increase from 2020 to 2023. This could be for many reasons, one being that there are more ways and availability to report cases. This ability helps people find easier access and is explored in another question in our data in rising user report cases by submission types.
